```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
air_bnb = pd.read_csv('../day-1/files/AB_NYC_2019.csv')
air_bnb.head(1)
```

```python
# How many neighborhood groups are available and which shows up the most?

Nbrhood_counts = air_bnb['neighbourhood_group'].value_counts()
Nbrhood_counts
```

```python
# Are private rooms the most popular in manhattan?

nbrhood_room_type=air_bnb.groupby(['neighbourhood_group', 'room_type']).size()
nbrhood_room_type
```

```python
# Which hosts are the busiest based on their reviews?

reviews = air_bnb[['host_name', 'reviews_per_month']]
sorted_data = reviews.sort_values('reviews_per_month', ascending = False).reset_index()
sorted_data.head()
```

```python
# Which neighorhood group has the highest average price?

air_bnb.groupby('neighbourhood_group').mean()[['price']].sort_values('price', ascending = False).head(1)
```

```python
# Which neighborhood group has the highest total price?
air_bnb[['neighbourhood_group','price']].groupby('neighbourhood_group').sum(['price']).sort_values('price', ascending = False).head(1)
```

```python
# Which top 5 hosts have the highest total price?
air_bnb[['host_id','price']].groupby(by="host_id").sum(['price']).sort_values('price', ascending = False).head()
```

```python
# Who currently has no (zero) availability with a review count of 100 or more?

no_availability=air_bnb[air_bnb['availability_365']==0]
no_availability
hundred_reviews_plus_no_avail=no_availability[no_availability['number_of_reviews']>=100].sort_values('number_of_reviews', ascending = False)
hundred_reviews_plus_no_avail
```

```python
# What host has the highest total of prices and where are they located?

air_bnb[['host_name','price', 'neighbourhood_group']].groupby('host_name').sum(['price']).sort_values('price', ascending = False).head()                                                      
                                                              
```

```python
# When did Danielle from Queens last receive a review?

danielle_queens = air_bnb[air_bnb['neighbourhood_group']=='Queens']
danielle = air_bnb_queens[danielle_queens['host_name']=='Danielle']
danielle['last_review'].sort_values()
```

## Further Questions

1. Which host has the most listings?

```python
air_bnb[['host_id', 'id']].groupby(['host_id']).count().sort_values('id', ascending = False).head(1)
```

2. How many listings have completely open availability?

```python
air_bnb[['availability_365']][air_bnb['availability_365']==365].count()
```

3. What room_types have the highest review numbers?

```python
air_bnb[['room_type', 'number_of_reviews']].groupby(['room_type']).sum().sort_values('number_of_reviews', ascending = False)
```
